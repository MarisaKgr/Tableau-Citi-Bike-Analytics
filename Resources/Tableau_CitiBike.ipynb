```python
import numpy as np
import pandas as pd
```

```python
#import csvs for aggregating
oct_2013 = pd.read_csv("2013-10 - Citi Bike trip data.csv")
oct_2014 = pd.read_csv("201410-citibike-tripdata.csv")
oct_2015 = pd.read_csv("JC-201510-citibike-tripdata.csv")
oct_2016 = pd.read_csv("JC-201610-citibike-tripdata.csv")
oct_2017 = pd.read_csv("JC-201710-citibike-tripdata.csv")
oct_2018 = pd.read_csv("JC-201810-citibike-tripdata.csv")
oct_2019 = pd.read_csv("JC-201910-citibike-tripdata.csv")
oct_2020 = pd.read_csv("JC-202010-citibike-tripdata.csv")
```

```python
#view dataframes
#oct_2013
#oct_2014 
#oct_2015
#oct_2016
#oct_2017
#oct_2018
#oct_2019
#oct_2020
```

```python
#concat to combine into one dataframe
october_2013_to_2020 = pd.concat([oct_2013,oct_2014,oct_2015,oct_2016,oct_2017,oct_2018,oct_2019,oct_2020])
october_2013_to_2020
```

```python
#replace the gender category values 0= unknown, 1= male, 2=female
october_2013_to_2020['gender']=october_2013_to_2020['gender'].astype(str)
october_2013_to_2020['gender'] = october_2013_to_2020['gender'].replace(['0','1','2'],['unknown','male','female'])
october_2013_to_2020
```

```python
#save csv for visualizations
october_2013_to_2020.to_csv("october_2013_to_2020.csv", encoding='utf-8', index=True)
```
